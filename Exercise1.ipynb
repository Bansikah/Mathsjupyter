```python
#!pip install numpy
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.linspace(-10, 10, 100)  # Generate 100 x-values from -10 to 10
y = np.sin(x**2 + 2*x + 5)     # Calculate y-values for each x
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x^2 + 2x + 5)')
plt.title('Graph of sin(x^2 + 2x + 5)')
plt.grid(True)
plt.show()
```

```python
#####3. Basic Quadratic Function: Implement and plot $g(x) = x^2 + 2x + 5$.
```

```python
def g(x):
    return x**2 + 2*x + 5
```

```python
x = np.linspace(-10, 10, 100)
```

```python
y = g(x)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Graph of g(x) = x^2 + 2x + 5')
plt.grid(True)
plt.show()
```

```python
####4. Sine Function: Implement the basic sine function $h(x) = \sin(x)$ and visualize it.
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('h(x)')
plt.title('Graph of h(x) = sin(x)')
plt.grid(True)
plt.show()
```

```python
#####5. Combine Quadratic & Sine: Implement $i(x) = \sin(x^2 + 2x + 5)$ and visualize.
```

```python
x = np.linspace(-10, 10, 100)
```

```python
y = np.sin(x**2 + 2*x + 5)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('i(x)')
plt.title('Graph of i(x) = sin(x^2 + 2x + 5)')
plt.grid(True)
plt.show()
```

```python
####5. Exponential Decay: Visualize the function $j(x) = e^{-0.1x}$ using the exponential function.
```

```python
####7.Complete Function: Combine all the above components to implement and visualize $f(x) = \sin(x^2 + 2x + 5)$ times $e^{-0.1x}$.
```

```python
x = np.linspace(-10, 10, 100)
```

```python
y = np.sin(x**2 + 2*x + 5) * np.exp(-0.1 * x)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = sin(x^2 + 2x + 5) * e^(-0.1x)')
plt.grid(True)
plt.show()
```

```python
# ######Certainly! Let's compare and comment on the graphs that have been drawn using Jupyter Notebook:

# 1. Quadratic Function ($g(x) = x^2 + 2x + 5$):
#    - The graph of the quadratic function forms a parabola.
#    - It is symmetric around the vertex, which is the minimum point of the parabola.
#    - The vertex of the parabola can be found using the formula $x = -\frac{b}{2a}$.
#    - The graph opens upward when the coefficient of $x^2$ is positive (as in this case).

# 2. Sine Function ($h(x) = \sin(x)$):
#    - The graph of the sine function is a periodic wave.
#    - It oscillates between -1 and 1 as the input values increase.
#    - The period of the wave is $2\pi$, meaning it repeats every $2\pi$ units.
#    - The amplitude of the wave is 1, which represents the maximum value.

# 3. Combined Quadratic and Sine Function ($i(x) = \sin(x^2 + 2x + 5)$):
#    - The graph combines the properties of the quadratic and sine functions.
#    - The quadratic term $(x^2 + 2x + 5)$ modifies the input to the sine function.
#    - It creates variations in the amplitude and frequency of the sine wave.
#    - The resulting graph is a combination of oscillations and quadratic shape.

# 4. Exponential Decay Function ($j(x) = e^{-0.1x}$):
#    - The graph of the exponential decay function is a decreasing curve.
#    - It starts at 1 and approaches 0 as the input values increase.
#    - The decay rate is controlled by the coefficient of $x$ (in this case, -0.1).
#    - The graph exhibits rapid decay initially and slows down as $x$ increases.

# 5. Combined Quadratic, Sine, and Exponential Decay Function ($f(x) = \sin(x^2 + 2x + 5) \times e^{-0.1x}$):
#    - The graph combines the properties of the quadratic, sine, and exponential functions.
#    - The quadratic term modifies the input to the sine function, creating variations in amplitude and frequency.
#    - The exponential decay term influences the overall decay of the function.
#    - The resulting graph is a combination of oscillations, quadratic shape, and decay.

# Each graph has its unique characteristics and behaviors based on the underlying functions. Understanding these properties helps in interpreting and analyzing the data they represent.
```
